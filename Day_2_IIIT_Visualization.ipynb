Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
 #set the color codes based on the function we use
 sns.set(color_codes=True)
```

```python
# randomly generate numbers for plots, 100 random numbers in the range 0-1
x=np.random.random(100)
## make 100 normally generated, make the function with keyword so that no need to worry about the order of the argument
y=np.random.normal(size=100)

```

```python
##print first 10 values
print(x[:10])
print(y[:10])
```

#

```python
# frequency distribution of data in a given range
plt.hist(x)
```

```python
# increase bins
plt.hist(x, bins=100)
```

```python
# increase bins, as y is normally distributed, getting bell shaped curve
plt.hist(y, bins=20)
```

```python
# increase bins, as y is normally distributed, getting bell shaped curve, coloured
plt.hist(y, bins=20, color="red")
```

```python
## With sns
sns.histplot(x)
```

```python
## With sns, and also draw a curve, kernel distribution (kde), as it is normally distrinuted, curver bell shape
sns.histplot(y, bins=20,kde=True)
```

```python
## With sns, and also draw a curve, kernel distribution (kde), as x is not normally distrinuted, curver bell shape
sns.histplot(x, bins=20,kde=True)
```

```python
## Plot only curve means kdeplot, also fill the curve
sns.kdeplot(y,color="magenta", fill=True)
```

```python
### To combine both the diagrams in single plot
sns.kdeplot(x,color="red", fill=True)
sns.kdeplot(y,color="magenta", fill=True)
```

```python
# single row, 2 column
plt.subplot(1,2,1)
sns.kdeplot(y,color="magenta", fill=True)
plt.subplot(1,2,2)
sns.kdeplot(x,color="red", fill=True)
```

```python
# 2 rows, 1 column
plt.subplot(2,1,1)
sns.kdeplot(y,color="magenta", fill=True)
plt.subplot(2,1,2)
sns.kdeplot(x,color="red", fill=True)
```

```python
# single row, 2 column, histplot
plt.subplot(1,2,1)
sns.histplot(y,color="green", fill=True)
plt.subplot(1,2,2)
sns.histplot(x,color="blue", fill=True)
```

```python
## To know "daywise", we go for barPlot than histPlot

days=tips_df.day.unique()
print(days)

### we need a dataframe, so we use dict() as dict() can be converted to dataframe directly, after converting to list
day_data={}

### Find the customers each day
for each_day in days:
  day_data[each_day]=[tips_df[tips_df["day"]==each_day]["day"].count()] #list because datframe only takes list

print(day_data)
```

```python
### convert the dict "data_data" to dataframe
day_df=pd.DataFrame(day_data)

## barplot
sns.barplot(day_df)
```

```python
# Grup by "day"
tips_df.groupby("day").count()
```

```python
# Grup by "day", available as a series
tips_series=tips_df.groupby("day")["day"].count()

print(tips_series)

sns.barplot(x=tips_series.index,y=tips_series.values)
```

```python
# Grup by "Sex", available as a series, any categorical group we can plot
tips_series=tips_df.groupby("sex")["day"].count()

print(tips_series)

sns.barplot(x=tips_series.index,y=tips_series.values)
```

```python
### Relational plots with 2 or more : 3 min things need to give, data, x and y
sns.relplot(data=tips_df,x="total_bill",y="tip")
## as total bill increases, amount of tip increases
```

```python
# give regression line, it will give a line
sns.regplot(data=tips_df,x="total_bill",y="tip")
## as total bill increases, amount of tip increases
```

```python
## linear model plot, more flexible

sns.lmplot(data=tips_df,x="total_bill",y="tip")
## as total bill increases, amount of tip increases, looks similar as regplot, but different options
```

```python
### some relationships we can take
sns.lmplot(data=tips_df[tips_df.sex=="Male"],x="total_bill",y="tip")
## as total bill increases, amount of tip increases
```

```python
### Color change
sns.lmplot(data=tips_df,x="total_bill",y="tip", hue='sex')

## male, one color, female, different color; divied based on sex
```

```python
### Color change
sns.lmplot(data=tips_df,x="total_bill",y="tip", hue='day')
```

```python
### based on smoking habits and sex, in columns we need to plot smoker and on rows sex  we need to observe
### Color change
sns.lmplot(data=tips_df,x="total_bill",y="tip", hue='day', col="smoker",row='sex')
```

```python
### we can visualise all attributes in pairplot, only on numerical data, rowise and columnwise it is giving histogram
sns.pairplot(tips_df)
## Explore all the plots and then check each plot one by one in more detail
```

```python
## pie chart not preferrable, as it takes a lot of time,
## but one more chart is there donut chart,
## to convert that into donut chart, mention wedgeprops
## Wedge is nothing but edge, we kept widths as 0.25

## separate components with explode


### Importance of donut chart compared to pie chart is , inside the chart we can have other circles, (like percentage of dinner etc), much flexibility

a=plt.pie(tips_series,labels=tips_series.index,autopct="%0.2f%%",shadow=True, wedgeprops=dict(width=0.25),explode=[0,1,0,0])
plt.savefig("test.svg")
```
